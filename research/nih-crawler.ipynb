```python
from llama_index.readers.web import WholeSiteReader


# Initialize the scraper with a prefix URL and maximum depth
scraper = WholeSiteReader(
    prefix="https://www.nia.nih.gov/health/alzheimers", max_depth=10  # Example prefix
)

# Start scraping from a base URL
documents = scraper.load_data(
    base_url="https://www.nia.nih.gov/health.html"
)  # Example base URL
```

```python
documents
```

```python
documents = [
    doc for doc in documents if not doc.text.startswith("403 ERROR")
]  # 403 ERROR due to cloudflare protection
for doc in documents[1:28]:
    content = doc.to_json()
    with open(f"../data/nih/alzheimers/{doc.doc_id}.json", "w") as f:
        f.write(content)
```
